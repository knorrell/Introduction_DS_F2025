```python
#Q3
s = "Intro to Data"

char_list = list(s)

print("s:", s, type(s))
print("char_list:", char_list, type(char_list))

first_4 = s[:4]
last_3 = s[-3:]
print("First 4 characters:", first_4)
print("Last 3 characters:", last_3)

char_list.append("Python")
print("Updated char_list:", char_list)
```

```python
#Q4

n = int(input("Enter an initial number: "))

while n <= 200:
    if n == 40:
        n *= 2
        continue 
    if n == 160:
        print("Stopping loop at value 160.")
        break
    print("Current value:", n)
    n *= 2
```

```python
#Q5

import numpy as np

arr = np.arange(1, 21)
print("Original array:", arr)

print("First 5 elements:", arr[:5])

print("Reversed array:", arr[::-1])

arr[arr % 2 == 0] = 0
print("Array with even numbers replaced by 0:", arr)
```
