```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('life_expectancy_cleaned.csv', index_col=0)
```

```python
# Write your code here
plt.figure(figsize=(8, 4))
plt.hist(df['Life_Expectancy_World_Bank'], bins=15, color='skyblue', edgecolor='black')
#OR: #df['Age'].plot(kind='hist', bins=15, color='skyblue', edgecolor='black')
plt.title('Univariate Analysis: Age Distribution')
plt.xlabel('Life_Expectancy_World_Bank')
plt.ylabel('Frequency')
plt.grid(True)
plt.savefig('life_expect_distr.jpg')
plt.show()
```

This histogram reveals the life expectancy across the world. You can see that the highest frequencies lie after the age of 70 indicating that most countries have a life expectancy beyond 70.

```python
corr_matrix= df.corr(numeric_only=True)

plt.figure(figsize=(9,6))
ax = sns.heatmap(
    corr_matrix, annot=True, fmt=".2f",
    cmap="coolwarm", center=0,
    linewidths=0.5, cbar_kws={"label": "Correlation"}
)
plt.title("Correlation Heat Map of Socioeconomic Factors")
plt.tight_layout()
plt.savefig('heatmap.jpg')
plt.show()
```

The correlation heat map illustrates the relationships between various socioeconomic and health-related factors. Darker red shades represent strong positive correlations, while deeper blue shades represent strong negative correlations. From the visualization, several notable patterns emerge.

One key finding is the strong positive correlation between CO2 emissions and Non-Communicable Diseases (approximately 0.82). This suggests that higher CO2 levels are associated with higher rates of non-communicable health conditions. Additionally, Injuries and Non-Communicable Diseases show a very strong positive correlation (0.95), indicating that regions with higher injury rates also tend to have higher rates of chronic diseases. Another important relationship appears between Life Expectancy (World Bank) and Above Average Life Expectancy, which show a strong positive correlation (0.83), confirming that these two measures of longevity move closely together.

Overall, the heat map helps highlight which factors tend to rise or fall together.
