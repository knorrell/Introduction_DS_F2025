## Introduction to Pandas
- Used for data manipulations and analysis
- two types of data (series: one dimension; data frame: two dimensional)

```python
import pandas as pd
list1=["Alice","Kaylah","Jordan","Klein"]
print(list1)
list2=[30,45,50,35]
print(list2)
```

```python
s=pd.Series([10,20,30,40])
print(s)
```

```python
dict_1={"Name":list1,"Score":list2}
df=pd.DataFrame(dict_1)
print(df)
```

```python
df.index=['a','b','c','d']
```
