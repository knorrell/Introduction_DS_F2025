```python
# Import Libraries
import numpy as np
import pandas as pd
```

```python
# Load the Dataset
df = pd.read_csv('Titanic_Dataset.csv')
df
```

```python
#Percentage of missing values
missing_data = df.isnull().sum()/len(df)*100 
print(missing_data)
df_new=df.copy()
```

```python
df_new['Pclass']=df_new['Pclass'].astype('category')
df_new['Pclass']=df_new['Pclass'].map ({1:'First', 2: 'Second', 3: 'Third'})
# Fill missing Age values based on the median age for each Pclass and Sex group
df_new['Age'] = df_new.groupby(['Pclass', 'Sex'],observed=True)['Age'].transform(lambda x: x.fillna(x.median()))
df_filtered = df_new.copy()
df_filtered.drop(columns=['Embarked','Cabin'],inplace=True)
print("\n Data Frame after dropping categorical column embarked & Cabin: \n")
df_filtered
```

```python
def classify_age(age):
    if pd.isnull(age):
        return 'Unknown'
    elif age < 13:
        return 'Child'
    elif age < 18:
        return 'Teen'
    elif age < 35:
        return 'Young Adult'
    elif age < 60:
        return 'Adult'
    else:
        return 'Senior'
df_filtered['AgeGroup'] = df_filtered['Age'].apply(classify_age)

df_filtered['FamilySize'] = df_filtered['SibSp'] + df_filtered['Parch'] + 1
#Adds up siblings/spouses and parents/children on board, plus 1 for the passenger.

df_filtered['IsAlone'] = (df_filtered['FamilySize'] == 1).astype(int)
#Creates a binary flag — 1 if the passenger had no family aboard, otherwise 0.

df_filtered
```
