```python
import numpy as np
M1=np.array([1,2,3])
M2=np.array([4,5,6])
print(f"Array M1:{M1}\n")
print(f"Array M2:{M2}\n")
M=np.concatenate((M1,M2))
print(f"Concatenated Array: {M}")
```

```python
a=np.array([1,2],
           [4,5],
           [6,7])
b=np.array([2,4],
           [1,5],
           [0,6])
C=np.concatenate((a,b))
print({C})
```

```python
a=np.array([[1,2],[3,4]])
b=np.array([[5,6],[7,8]])
vertical_concat=np.concatenate((a,b), axis=0)
print("Vertical Concatenation:")
print(vertical_concat)
```
