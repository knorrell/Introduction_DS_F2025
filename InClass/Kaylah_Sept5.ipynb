# Nested Dictionary

```python
# Dictionary with names as keys and subject-score pairs as values
scores = {
    "Alice": {"Math": 85, "Science": 90, "English": 78},
    "Bob": {"Math": 92, "Science": 88, "English": 81},
    "Charlie": {"Math": 76, "Science": 95, "English": 89}
}

print(scores)
```
