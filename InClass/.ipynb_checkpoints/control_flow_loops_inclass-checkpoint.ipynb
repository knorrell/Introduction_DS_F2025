### for loop in python

**Syntax** <br>
```
for variable in sequence:
	# code block to run

```

**Loop through a list**

````python
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)
````

**Loop through a string**

````python
for letter in "Data":
    print(letter)
````

**Loop through a `range()`**

````python
for i in range(5):
    print(i)
````

**Custom range (start, stop, step)**

````python
for i in range(2, 11, 2):
    print(i)
````

**Nested loop (loop inside a loop)**

````python
for i in range(3):       # outer loop
    for j in range(2):   # inner loop
        print(i, j)
````

**Examples:**

* Write a for loop to find the sum of numbers from 1 to 5. Print the total at the end.
* Write a for loop to print the square of numbers from 1 to 5.
* Write a nested for loop to print a multiplication table from 1 to 3.

### while loop in a python

**Syntax**:

```
while condition:
	# code block to run
```

**Counting**

````python
i = 1
while i <= 5:
    print(i)
    #i += 1
    i=i+1
````

**Loop until user input**

````python
password = ""
while password != "python123":
    password = input("Enter password: ")

print("Access granted!")
````
