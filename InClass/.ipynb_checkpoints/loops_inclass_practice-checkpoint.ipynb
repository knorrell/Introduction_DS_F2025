### for loop in python

**Syntax** <br>
```
for variable in sequence:
	# code block to run

```

---
**Example:** Write a for loop to print the square of numbers from 1 to 5.

---
**Example:** Write a for loop to find the sum of numbers from 1 to 5. Print the total at the end.

### while loop in a python

**Syntax**:

```
while condition:
	# code block to run
```

---
**Example:** Write a while loop that prints numbers from 5 down to 1.

---
**Example:**
Use a while loop to reverse the string "Python".
