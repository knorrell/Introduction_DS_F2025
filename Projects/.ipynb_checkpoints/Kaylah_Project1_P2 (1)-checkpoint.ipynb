Q2.

```python
print("Q2 Output:")
n = int(input("Enter an initial number: "))
while n<=200:
    if n ==40:
        n*=2
        continue
    if n == 160:
        print("Loop stopped because value became 160.")
        break
    print(f"Current value of n = {n}")
    n *= 2
print()
```

```python
import numpy as np
print("Q4 Output:")
arr = np.arange(1, 21)
print(f"Original array = {arr}")
print(f"First 5 elements = {arr[:5]}")
print(f"Array in reverse order = {arr[::-1]}")

arr[arr % 2 == 0] = 0
print(f"Array with even numbers replaced by 0 = {arr}")
print()
```

Q5

```python
print("Q5 Output:")

A = np.array([[0,1],
              [5,2],
              [8,9]])

B = np.array([[3,5],
              [2,4],
              [1,3]])

print(f"Matrix A =\n{A}")
print(f"Matrix B =\n{B}")

print(f"Shape of A = {A.shape}")
print(f"Dimension of A = {A.ndim}")

print(f"Shape of B = {B.shape}")
print(f"Dimension of B = {B.ndim}")

print(f"Element at row 3, column 2 of B =", B[2, 1])
print(f"First column of A =\n{A[:, 0]}")

sum_matrices = A + B
sub_matrices = A - B
mult_matrices = A * B

print(f"Sum of matrices =\n{sum_matrices}")
print(f"Subtraction of matrices =\n{sub_matrices}")
print(f"Element-wise multiplication of matrices =\n{mult_matrices}")
print()
```
